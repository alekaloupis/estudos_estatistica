```python
#Imports 
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom
import numpy as np
import seaborn as sns
```

```python
#Função que implementa uma distribuição binomial
#Entradas
# n - numero total de tentativas ou ensaios;
# p - probabilidade de sucesso com casas decimais;
#Saida
# df - dataframe com numero de sucessos desejados e lista de probabilidades
# p - probabilidade de sucesso
def Binomial_Dist(n,p):
    r = list (range(0,n+1))
    list_probabilities = 100 *(np.round(binom.pmf(r,n,p),3))
    data = {'numero_sucessos': r,'probabilidades':list_probabilities}
    df = pd.DataFrame(data = data)
    return df, p
```

```python
#Funcao que plota os sucessos e as probabilidades para as diferentes distribuicoes geradas
#Entradas: Dataframe1 e Dataframe2 + probabilidades de sucesso para distribuições binomiais 1 e 2
def Plot_Sucesso_Probabilidades(df1, p1, df2, p2):
    fig, ax = plt.subplots(figsize = (15,6))
    ax.set_axis_off()
    plt.subplot(1,2,1)
    plt.title("Numero de sucessos desejados x Probabilidades (dado p) =  {:.1%}".format(p1))
    sns.scatterplot(x = 'numero_sucessos', y = 'probabilidades', data = df1);
    plt.subplot(1,2,2)
    plt.title("Numero de sucessos desejados x Probabilidades (dado p) = {:.1%}".format(p2))
    sns.scatterplot(x = 'numero_sucessos', y = 'probabilidades', data = df2, color = 'orange');
```

#### Apontamento 1: É interessante observar que nas distribuições com probabilidade de sucesso maiores: à medida que o numero de sucessos desejados (r) aumenta, a probabilidade de sucesso também aumenta

#### Apontamento 2: Inversamente, as distribuições binomiais com probabilidade de sucesso menor, à medida que o numero de sucessos desejados (r) aumenta, a probabilidade de sucesso diminui

#### Os plots das distribuições com valores complementares uma ao lado da outra serve para demonstrarmos visualmente o quê foi descrito acima

```python
df1, p1 =  Binomial_Dist(100,0.95)
df2, p2 = Binomial_Dist(100,0.05)
Plot_Sucesso_Probabilidades(df1,p1, df2, p2)
```

```python
df1, p1 =  Binomial_Dist(1000,0.95)
df2, p2 = Binomial_Dist(1000,0.05)
Plot_Sucesso_Probabilidades(df1,p1, df2, p2)
```

```python
df1, p1 =  Binomial_Dist(50,0.95)
df2, p2 = Binomial_Dist(50,0.05)
Plot_Sucesso_Probabilidades(df1,p1, df2, p2)
```
