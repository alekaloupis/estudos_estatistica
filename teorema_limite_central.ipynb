```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Vamos plotar uma distribuicao com 20 valores
#Minha distribuicao tera muito mais valores de 1 e 6 do que de 2,3,4,5
a = [ y for y in range(1,2,1) for x in range(1,9,1)]
b = [ y for y in range(5,6,1) for x in range(1,9,1)]
c = [2,3,4,6]
lista_final = a + b + c 
lista_final
```

```python
def Normal_Dist_Plot(lista_final, n):
    #Função para receber a lista de valores e a quantidade de medias amostrais que desejamos calcular e plotar a distribuicao
    # dessas medias
    
    #Entradas:
    # lista_final = lista com 20 valores, variando entre 1 e 6, porem com predominancia dos valores 1 e 6
    # n = quantidade de medias amostrais que serão calculadas
    
    #Saida:
    #Plot com a distribuicao das medias
    final_mean = []
    for x in range(0,n,1):
        aux = [np.random.choice(lista_final) for x in range(0,20,1)]
        final_mean.append(np.mean(aux))
        
    df = pd.DataFrame(data= final_mean, columns=['final_mean'])
    data = df['final_mean'].value_counts().reset_index()
    aux1 = pd.DataFrame(data = data)
    fig = plt.figure(figsize=(15,10))
    sns.barplot(x = 'final_mean', y = 'count', data = aux1)
    plt.xticks(rotation = 45)
    plt.show()
```

```python
Normal_Dist_Plot(lista_final,40)
```
